```python
#filehandling in python
#open a file in write mode
file = open("example.txt", "w") # write mode - creates a new file or overwrites an existing file
file.write("Hello, this is a file handling example in Python.") #write some text to the file
file.close() # close the file

#open the file in read mode
file = open("example.txt", "r") # read mode - opens the file for reading
content = file.read() # read the entire content of the file
print(content) # print the content
file.close() # close the file

#open the file in append mode
file = open("example.txt", "a") # append mode - opens the file for writing and appends to the end of the file
file.write("\nThis line is appended to the file.") # append some text to the file   
file.close() # close the file

#creating a new file if file does not exist, return error if file already exists
file = open("new_example.txt", "x") # create mode - creates a new file or raises an error if the file already exists
file.write("Hello, this is a new file.") # write some text to the file
file.close() # close the file
```
