```python
#controlflow
#if, else, elif
age = int(input("enter your age: "))
if age >= 18:
    print("you are eligible to vote")
else:
    print("you are not eligible to vote")
```

```python
#elif
marks = int(input("enter your marks: "))
if marks >= 80:
    print("you got an A grade")
elif marks >= 60:
    print("you got a B grade")
else:
    print("you got a C grade")
```
