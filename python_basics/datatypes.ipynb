| Type           | Syntax        | Ordered | Changeable | Allows Duplicates    | Indexable              | Example                      |
| -------------- | ------------- | ------- | ---------- | -------------------- | ---------------------- | ---------------------------- |
| **List**       | `[ ]`         | ✅ Yes   | ✅ Yes      | ✅ Yes                | ✅ Yes                  | `["apple","banana","apple"]` |
| **Tuple**      | `( )`         | ✅ Yes   | ❌ No       | ✅ Yes                | ✅ Yes                  | `("apple","banana")`         |
| **Set**        | `{ }`         | ❌ No    | ✅ Yes      | ❌ No                 | ❌ No                   | `{"apple","banana"}`         |
| **Dictionary** | `{key:value}` | ✅ Yes   | ✅ Yes      | ❌ Keys cannot repeat | ❌ No *(access by key)* | `{"name":"John","age":25}`   |


```python
#Dictionary
person = {
    "name": "pradeep",
    "age": 30,
    "city": "chennai"
}   
print(person)
print(person["name"])
print(person["age"])
print(person["city"])
```

```python
#set
numbers = {1, 2, 3, 4, 5}
print(numbers)
print(numbers)
fruits = {"apple", "banana", "orange", "apple"}  # Duplicate will be ignored
print(fruits)
```

```python
#tuple
numbers = (1, 2, 3, 4, 5)
print(numbers)
# numbers[0] = 10  # This will raise an error because tuples are immutable
# print(numbers)

print(numbers[0])
```

```python
#list
numbers = [1, 2, 3, 4, 5]
print(numbers)
print(numbers[0])
print(numbers[1])
print(numbers[2])
print(numbers[3])
print(numbers[4])
```
