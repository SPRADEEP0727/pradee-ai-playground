```python
#Python loops
#while loop
i = 1
while i <= 5:
    print(i)
    i += 2
```

```python
#for loop
fruits = ["apple", "banana", "orange"] #apple=0 ; banana=1 ; orange=2
for fruit in fruits:
    print(fruit)
    
```

```python
#for loop with range
for i in range(1, 6): #range (start, end) 6 wont be included because it is exclusive -1
    print(i)
```
